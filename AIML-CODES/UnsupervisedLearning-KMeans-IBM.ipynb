```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# **Creating Random Data**

```python
data = np.random.rand(100,2)
```

```python
data.shape
```

```python
print(data)
```

# **Model Initialization**

```python
kmeans = KMeans(n_clusters = 3)

# Model Fitting
kmeans.fit(data)
```

```python
centroids = kmeans.cluster_centers_ # choosing k random points from dataset for centroid

labels = kmeans.labels_ # Labelling the clusters

print("Centers=\n", centroids)
print("Labels=\n", labels)
```

# **Using matplotlib to visualize**

```python
# plotting data points
plt.scatter(data[:,0], data[:,1], c=labels, cmap='viridis', edgecolors='k')

# plotting centroids
plt.scatter(centroids[:,0], centroids[:,1], marker = 'x', c='red', s=100, label='Centroids')

# 
plt.title("K-Means Cluster:")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show
```
