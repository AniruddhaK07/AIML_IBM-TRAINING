# **Importing Libraries**

- File reading - Pandas
- Data Manipulation - Numpy

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, mean_squared_error, r2_score

print("Libraries Imported Sucessfully.")
```

# **Reading Files**

```python
data = pd.read_csv("datasets/loan_data.csv")
print("Data read successfully.")
```

# **Separating Numerical & Categorical columns**

```python
num_col = data.select_dtypes(include=['int64','float64']).columns
cat_col = data.select_dtypes(include=['object']).columns

print("Numerical -", list(num_col))
print("Categorical -", list(cat_col)) 
```

# **Encoding Categorical Columns**

```python
data_encoded = data.copy()
for col in cat_col:
    data_encoded[col] = data_encoded[col].astype('category').cat.codes
```

```python
data_encoded.head()
```

# **Splitting data into feature and label**

```python
# subtle error handling
# outputs
target_class = 'loan_status' if ' loan_status' in data_encoded.columns else data_encoded.columns[-1]
target_regr = 'ApplicationIncome' if 'ApplicationIncome' in data_encoded.columns else num_col[-1]

# inputs
x_class = data_encoded.drop(columns=[target_class]) # input for classification
y_class = data_encoded[target_class] # output for classification

x_regr = data_encoded.drop(columns=[target_regr]) # input for regressor
y_regr = data_encoded[target_regr] # output for regressor
```

# **Data Validation**

```python
# for classification
xc_train, xc_test, yc_train, yc_test = train_test_split(x_class, y_class, test_size=0.2, random_state=42)

# for regressor
xr_train, xr_test, yr_train, yr_test = train_test_split(x_regr, y_regr, test_size=0.2, random_state=42)
```

# **Decision Tree**

```python
# 1. Classification

dt_clf = DecisionTreeClassifier(random_state=42)

dt_clf.fit(xc_train, yc_train)

yc_pred_dt = dt_clf.predict(xc_test)
```

```python
# 2. Regression

dt_clf = DecisionTreeRegressor(random_state=42)

dt_clf.fit(xr_train, yr_train)

yr_pred_dt = dt_clf.predict(xr_test)
```

# **Random Forest**

1. Classification

```python
rf_clf = RandomForestClassifier(random_state=42)

rf_clf.fit(xc_train, yc_train)

yc_pred_rf = rf_clf.predict(xc_test)
```

# **Evaluation**

```python
print("==================Classification Results=================")
print("\n DT Accuracy \n", accuracy_score(yc_test,yc_pred_dt))
print("\n RF Accuracy \n", accuracy_score(yr_test,yr_pred_rf))

print("\n Decision Tree Classification Report\n", classification_report(yc_test, yc_pred_dt))
print("\n Random Forest Classification Report\n", classification_report(yc_test, yc_pred_rf))
```
